```python
finalx=Fframe.drop(columns=["ptsd","PTSD_Risk",'gender',"age","education","household_income"],axis="columns")
finaly=Fframe["ptsd"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(finalx,finaly,train_size=.8,random_state=42)
y_test.tail()
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_test,y_test)
```

```python
lg.predict(x_test)
```

```python
lg.score(x_test,y_test)
```

```python
lg.predict([[3,1,1,0,0,1,1,1]])
```

```python
metrics.confusion_matrix(y_test,lg.predict(x_test))
```

```python
from sklearn.svm import SVC
sv=SVC(kernel="linear")
sv.fit(x_train,y_train)
```

```python
sv.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

```python
from sklearn.linear_model import Perceptron
p=Perceptron()
p.fit(x_train,y_train)
```

```python
p.score(x_test,y_test)
```

```python
yp_lg=lg.decision_function(x_test)
yp_sv=sv.decision_function(x_test)
yp_rf=rf.decision_path(x_test)
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
lg_fpr,lg_tpr,threthold1=roc_curve(y_test,yp_lg)
sv_fpr,sv_tpr,threthold2=roc_curve(y_test,yp_sv)
#rf_fpr,rf_tpr,threthold3=roc_curve(y_test,yp_rf)
```

```python
auc_lg=auc(lg_fpr,lg_tpr)
auc_sv=auc(sv_fpr,sv_tpr)
```

```python
plt.figure(figsize=(8,6))
plt.plot(lg_fpr,lg_tpr, linestyle="dashed", label="logistic(auc=%.3f)"%auc_lg)
plt.plot(sv_fpr,sv_tpr, linestyle="dashed", label="svm(auc=%.3f)"%auc_sv)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("Receiver operating characteristic example")
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dataset=pd.DataFrame(iris.data,columns=iris.feature_names)
dataset["target"]=iris.target
dataset.head()
```

```python
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 4].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
#not necessary al
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```
